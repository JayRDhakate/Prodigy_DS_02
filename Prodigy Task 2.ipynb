# Handling Missing data

```python
df.isnull().sum()
```

```python
df1.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].median())
df1
```

```python
df['Embarked'] = df['Embarked'].fillna(method = 'ffill')
df
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df
```

```python
df1['Sex'] = df1['Sex'].map({'male': 0, 'female': 1})
df1
```

```python
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df
```

```python
df1['Embarked'] = df1['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df1
```

```python
#Dropping unnecessary columns
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df
```

```python
df1.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df1
```

# Checking for Correlation

```python
#correlation matrix of train data
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Correlation matrix of test data
corr_matrix_test = df1.corr()
sns.heatmap(corr_matrix_test, annot=True, cmap='coolwarm')
plt.show()
```

```python
#the survival rate by different features
print(df.groupby('Pclass')['Survived'].mean())
print(df.groupby('Sex')['Survived'].mean())
print(df.groupby('Embarked')['Survived'].mean())
```

```python
#age distribution among survivors and non-survivors
plt.figure(figsize=(10, 5))
sns.histplot(df[df['Survived'] == 1]['Age'], label='Survived', kde=True)
sns.histplot(df[df['Survived'] == 0]['Age'], label='Did not survive', kde=True)
plt.legend()
plt.show()
```
